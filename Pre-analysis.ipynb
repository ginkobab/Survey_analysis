```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = 10
pd.options.display.max_rows = 500
```

```python
data = pd.read_csv('clean_dati_def.csv')
```

```python
data.columns
```

```python
data['eta_1'] = data['eta_1'].fillna(99)
data['eta_1'] = data['eta_1'].astype(int)

plt.figure(figsize=(24,10))
sns.countplot(x='eta_1', data=data)
```

```python
data['altezza_1'] = data['altezza_1'].fillna(0)
data['altezza_1'] = data['altezza_1'].astype(int)


plt.figure(figsize=(24,10))
sns.distplot(data['altezza_1'])
```

```python
plt.figure(figsize=(3, 6))
sns.countplot(x='sesso', data=data)
```

```python
data['etnia_1'] = data['etnia_1'].fillna('None')

plt.figure(figsize=(20,40))
plot = sns.countplot(y='etnia_1', data=data, order = data['etnia_1'].value_counts(ascending=True).index)
# plot.set_xticklabels(plot.get_xticklabels(), rotation=90)

plt.show(plot)
```

```python
plt.figure(figsize=(24,10))
plt.xlim(0,1000)
sns.boxplot(data['Duration (in seconds)'])
```

```python
data['peso_1'] = data['peso_1'].fillna(0)
data['peso_1'] = data['peso_1'].astype(int)


plt.figure(figsize=(24,10))
sns.kdeplot(data['peso_1'], shade=True)
```

```python
plt.figure(figsize=(24,10))


sns.violinplot(x = 'etnia_1', y = 'peso', data= data.loc[data['etnia_1' == ["Stati Uniti d'America", 'Italia']]])

```

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

pd.options.display.max_columns = 150
%matplotlib inline
```

```python
data = pd.read_csv('/home/ginko/A-Projects/Questionario_alimentazione/dati_preanalisi/preanalisi.csv')
data.head()
```

```python
variabili_punteggio_cibi = ['cena_cibi_1',
                            'cena_cibi_2',
                            'cena_cibi_3',
                            'cena_cibi_4']


def somma_cibi(dataframe): 
    # Questa funzione potrebbe dare il punteggio e togliere tutte le variabili che non ci servono in un modello di regressione logistica
    dataframe['punteggio_aliment'] =  dataframe['cena_cibi_1']/3 # Ovviamente va fatto bene
    dataframe.drop()
    return dataframe
new_data = somma_cibi(data)
```

```python
corr = data.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
